```python
bcr = load_iris()
```

```python
x = pd.DataFrame(bcr.data, columns=bcr.feature_names)
y = pd.Series(bcr.target)
```

```python
df = x.copy()
df['target'] = y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].hist(x.iloc[:, 0], bins=20, color='blue', alpha =1)   # before scaling seple linght

ax[1].hist(x_train[:, 0], bins=20, color='green', alpha = 0.5)  # after scaling seple linght
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].hist(x.iloc[:, 1], bins=20, color='blue', alpha =1)   # before scaling seple widht

ax[1].hist(x_train[:, 1], bins=20, color='green', alpha = 0.5)  # after scaling seple widht
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors= 5)
knn.fit(x_train, y_train)
y_pred_knn = knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred_knn))
cm = confusion_matrix(y_test, y_pred_knn)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth= 3, random_state= 42)
dtc.fit(x_train,y_train)
y_pred_dtc = dtc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred_dtc))
cm = confusion_matrix(y_test, y_pred_dtc)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
```

```python
# prompt: ab mera ml model train ho gaya hai. meko iris ke panch dataset de jo iris_dataset me nahi

# Since the original iris_dataset (df) contains all 150 samples, we need to generate 5 new, unseen samples.
# These new samples will be outside of the data used for training and testing the model.
# We can create these samples by generating random values within the reasonable range of each feature in the original dataset.
# The target values for these new samples are unknown, as they are new data points.

# Get the min and max values for each feature from the original dataset
min_vals = df.drop('target', axis=1).min()
max_vals = df.drop('target', axis=1).max()

# Generate 5 new random data points for each feature within the observed range
new_samples = pd.DataFrame({
    col: np.random.uniform(min_vals[col], max_vals[col], 5)
    for col in df.drop('target', axis=1).columns
})

print("Five new Iris samples not in the original dataset:")
print(new_samples)
```

```python
# prompt: ab uper ke data set mese index 0 wale dataset ko mare ml model me dal aur pata kar ki wo kon si spiceas se blonge karta hai

# Select the first sample (index 0) from the new_samples DataFrame
sample_to_predict = new_samples.iloc[[0]]

# The model was trained on scaled data, so we need to scale the new sample as well.
# Use the same scaler object that was fitted on the training data (x_train)
scaled_sample = scaler.transform(sample_to_predict)

# Use the trained k-Nearest Neighbors (knn) model to predict the species of the sample
predicted_species_knn = knn.predict(scaled_sample)

# Use the trained Decision Tree Classifier (dtc) model to predict the species of the sample
predicted_species_dtc = dtc.predict(scaled_sample)

# The target values are 0, 1, and 2. We can use the target_names from the bcr dataset to map these to species names.
species_names = bcr.target_names

# Print the predicted species for the first sample using both models
print(f"\nPrediction using KNeighborsClassifier for the first new sample:")
print(f"Predicted species index: {predicted_species_knn[0]}")
print(f"Predicted species name: {species_names[predicted_species_knn[0]]}")

print(f"\nPrediction using DecisionTreeClassifier for the first new sample:")
print(f"Predicted species index: {predicted_species_dtc[0]}")
print(f"Predicted species name: {species_names[predicted_species_dtc[0]]}")
```

```python
# prompt: same code for another index 3 and 5

# Select the third sample (index 2) from the new_samples DataFrame
sample_to_predict_3 = new_samples.iloc[[2]]

# Scale the third new sample using the same scaler
scaled_sample_3 = scaler.transform(sample_to_predict_3)

# Predict the species of the third sample using both models
predicted_species_knn_3 = knn.predict(scaled_sample_3)
predicted_species_dtc_3 = dtc.predict(scaled_sample_3)

# Print the predicted species for the third sample
print(f"\nPrediction using KNeighborsClassifier for the third new sample:")
print(f"Predicted species index: {predicted_species_knn_3[0]}")
print(f"Predicted species name: {species_names[predicted_species_knn_3[0]]}")

print(f"\nPrediction using DecisionTreeClassifier for the third new sample:")
print(f"Predicted species index: {predicted_species_dtc_3[0]}")
print(f"Predicted species name: {species_names[predicted_species_dtc_3[0]]}")

# Select the fifth sample (index 4) from the new_samples DataFrame
sample_to_predict_5 = new_samples.iloc[[4]]

# Scale the fifth new sample using the same scaler
scaled_sample_5 = scaler.transform(sample_to_predict_5)

# Predict the species of the fifth sample using both models
predicted_species_knn_5 = knn.predict(scaled_sample_5)
predicted_species_dtc_5 = dtc.predict(scaled_sample_5)

# Print the predicted species for the fifth sample
print(f"\nPrediction using KNeighborsClassifier for the fifth new sample:")
print(f"Predicted species index: {predicted_species_knn_5[0]}")
print(f"Predicted species name: {species_names[predicted_species_knn_5[0]]}")

print(f"\nPrediction using DecisionTreeClassifier for the fifth new sample:")
print(f"Predicted species index: {predicted_species_dtc_5[0]}")
print(f"Predicted species name: {species_names[predicted_species_dtc_5[0]]}")
```

```python
# prompt: construct a graph of all iris dataset and new_sample dataset and find simmilarity

# To visualize the relationship between the original Iris dataset and the new samples,
# we can combine them into a single DataFrame for plotting.
# Since the new samples don't have a known 'target', we can assign a distinct value
# (e.g., -1) to represent them in the combined dataset for visualization purposes.

# Add a 'target' column to the new_samples DataFrame with a placeholder value
new_samples_vis = new_samples.copy()
new_samples_vis['target'] = -1

# Combine the original scaled training data and the scaled new samples
# Note: We are using the scaled training data (x_train) for comparison,
# as the new samples were also scaled using the same scaler.
# We'll also include the original test data (x_test) and their true labels (y_test)
# to see how the new samples relate to the known classes in the test set.

# Convert scaled x_train and x_test back to DataFrames with original column names
x_train_df = pd.DataFrame(x_train, columns=bcr.feature_names)
x_test_df = pd.DataFrame(x_test, columns=bcr.feature_names)

# Add the 'target' column to x_train and x_test dataframes
x_train_df['target'] = y_train.values
x_test_df['target'] = y_test.values

# Combine the three dataframes
combined_df = pd.concat([x_train_df, x_test_df, new_samples_vis], ignore_index=True)

# Map the target values to meaningful labels for plotting
target_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica', -1: 'new_sample'}
combined_df['species'] = combined_df['target'].map(target_map)

# Plotting the combined dataset
# Using pairplot to visualize pairwise relationships between features,
# highlighting the original species and the new samples.
plt.figure(figsize=(10, 8))
sns.pairplot(combined_df, hue='species', palette='viridis', diag_kind='kde', corner=True)
plt.suptitle('Iris Dataset with New Samples and Species')
plt.show()

# To find similarity, we can compute the Euclidean distance between each new sample
# and all samples in the original Iris dataset (training and test combined).
# A smaller distance indicates higher similarity.

# Combine original scaled data for distance calculation
original_scaled_data = np.vstack((x_train, x_test))
original_targets = np.concatenate((y_train.values, y_test.values))

# Calculate Euclidean distance from each new sample to all original data points
from sklearn.metrics.pairwise import euclidean_distances

distances = euclidean_distances(scaled_sample, original_scaled_data)
distances_3 = euclidean_distances(scaled_sample_3, original_scaled_data)
distances_5 = euclidean_distances(scaled_sample_5, original_scaled_data)


# Find the index of the closest original sample for each new sample
closest_index_1 = np.argmin(distances)
closest_index_3 = np.argmin(distances_3)
closest_index_5 = np.argmin(distances_5)

# Get the target (species) of the closest original sample
closest_species_1 = species_names[original_targets[closest_index_1]]
closest_species_3 = species_names[original_targets[closest_index_3]]
closest_species_5 = species_names[original_targets[closest_index_5]]

print("\nSimilarity Analysis:")
print(f"The first new sample is closest to an original sample of species: {closest_species_1}")
print(f"The third new sample is closest to an original sample of species: {closest_species_3}")
print(f"The fifth new sample is closest to an original sample of species: {closest_species_5}")

# You can also look at the distribution of distances to each class for more insight
# For example, for the first new sample:
distances_by_class = {
    species_name: distances[0][original_targets == i]
    for i, species_name in enumerate(species_names)
}

print("\nDistances of the first new sample to samples of each original species:")
for species, dists in distances_by_class.items():
    print(f"  {species}: Mean distance = {np.mean(dists):.4f}, Min distance = {np.min(dists):.4f}")

# This output shows that the first new sample is on average closest to the species
# identified by the minimum distance. The minimum distance itself confirms which single
# original sample is the absolute closest.
```

```python
# prompt: aisa graph bana ke mai khuud se iris data aur new sample ko compare karsako

# Function to visualize a new sample against the original data
def plot_sample_comparison(original_df, new_sample_df, sample_name, scaler, bcr_target_names):
    """
    Plots a new sample against the original Iris dataset using pairplot.

    Args:
        original_df (pd.DataFrame): The original Iris DataFrame (including 'target').
        new_sample_df (pd.DataFrame): A DataFrame containing the new sample(s).
        sample_name (str): A name for the new sample(s) for the plot legend.
        scaler: The StandardScaler object used for scaling.
        bcr_target_names (list): List of Iris species names.
    """
    # Create a copy to avoid modifying the original dataframe
    plot_df = original_df.copy()

    # Add a target column to the new sample(s) with a placeholder value
    new_sample_vis = new_sample_df.copy()
    new_sample_vis['target'] = -1 # Use -1 for new samples

    # Combine the original and new sample data
    combined_df = pd.concat([plot_df, new_sample_vis], ignore_index=True)

    # Map target values to species names for plotting
    target_map = {i: name for i, name in enumerate(bcr_target_names)}
    target_map[-1] = sample_name # Map the placeholder to the sample name
    combined_df['species'] = combined_df['target'].map(target_map)

    # Plotting the combined dataset
    plt.figure(figsize=(10, 8))
    sns.pairplot(combined_df, hue='species', palette='viridis', diag_kind='kde', corner=True)
    plt.suptitle(f'Iris Dataset vs. {sample_name}')
    plt.show()

# Example usage: Plotting the first new sample against the original dataset
plot_sample_comparison(df, new_samples.iloc[[0]], 'Sample 1', scaler, bcr.target_names)

# Example usage: Plotting the third new sample against the original dataset
plot_sample_comparison(df, new_samples.iloc[[2]], 'Sample 3', scaler, bcr.target_names)

# Example usage: Plotting the fifth new sample against the original dataset
plot_sample_comparison(df, new_samples.iloc[[4]], 'Sample 5', scaler, bcr.target_names)

# You can create a dropdown to select which new sample to plot interactively
# This requires the IPython display module
from IPython.display import display
from ipywidgets import Dropdown, Output

# Create a dropdown widget
sample_dropdown = Dropdown(
    options=[(f'Sample {i+1}', i) for i in range(len(new_samples))],
    description='Select Sample to Plot:',
)

# Create an output widget to display the plot
output_widget = Output()

# Function to call when the dropdown value changes
def on_dropdown_change(change):
    with output_widget:
        output_widget.clear_output(wait=True)
        selected_sample_index = change['new']
        selected_sample_df = new_samples.iloc[[selected_sample_index]]
        sample_name = f'Sample {selected_sample_index + 1}'
        plot_sample_comparison(df, selected_sample_df, sample_name, scaler, bcr.target_names)

# Observe changes in the dropdown value
sample_dropdown.observe(on_dropdown_change, names='value')

# Display the dropdown and the output area
display(sample_dropdown, output_widget)

# You can also add another dropdown to select which sample to perform similarity analysis on
similarity_sample_dropdown = Dropdown(
    options=[(f'Sample {i+1}', i) for i in range(len(new_samples))],
    description='Select Sample for Similarity:',
)

similarity_output_widget = Output()

# Function to perform and display similarity analysis
def on_similarity_dropdown_change(change):
    with similarity_output_widget:
        similarity_output_widget.clear_output(wait=True)
        selected_sample_index = change['new']
        selected_sample_df = new_samples.iloc[[selected_sample_index]]
        sample_name = f'Sample {selected_sample_index + 1}'

        # Scale the selected new sample
        scaled_selected_sample = scaler.transform(selected_sample_df)

        # Calculate Euclidean distance to all original scaled data
        distances = euclidean_distances(scaled_selected_sample, original_scaled_data)

        # Find the index of the closest original sample
        closest_index = np.argmin(distances)

        # Get the target (species) of the closest original sample
        closest_species = species_names[original_targets[closest_index]]

        print(f"\nSimilarity Analysis for {sample_name}:")
        print(f"This sample is closest to an original sample of species: {closest_species}")

        # Look at the distribution of distances to each class
        distances_by_class = {
            species_name: distances[0][original_targets == i]
            for i, species_name in enumerate(species_names)
        }

        print(f"\nDistances of {sample_name} to samples of each original species:")
        for species, dists in distances_by_class.items():
            print(f"  {species}: Mean distance = {np.mean(dists):.4f}, Min distance = {np.min(dists):.4f}")

# Observe changes in the similarity dropdown value
similarity_sample_dropdown.observe(on_similarity_dropdown_change, names='value')

# Display the similarity dropdown and output area
display(similarity_sample_dropdown, similarity_output_widget)
```
